# Function to Clean Weather Data

```python
# function to clean weather data
def clean_weather_data(file_path):
    """
    Cleans weather data from a CSV file.
    Args:
        file_path (str): Path to the CSV file.
    Returns:
        pandas.DataFrame: Cleaned DataFrame.
    """

    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return None

    # Select relevant columns
    selected_columns = ['days__datetime', 'days__tempmax', 'days__tempmin', 'days__temp',
                        'days__feelslikemax', 'days__feelslikemin', 'days__feelslike',
                        'days__precip', 'days__precipprob', 'days__precipcover',
                        'days__preciptype__001', 'days__preciptype__002', 'days__snow',
                        'days__snowdepth', 'days__windspeed', 'days__cloudcover',
                        'days__visibility']
    df_clean = df[selected_columns].copy()

    # Convert datetime column
    df_clean['days__datetime'] = pd.to_datetime(df_clean['days__datetime'])

    # Add day of the week
    df_clean['days_day_of_week'] = df_clean['days__datetime'].dt.day_name()

    # Create rain and snow columns
    df_clean['days_has_rain'] = df_clean.apply(lambda row: 1 if ('rain' in str(row['days__preciptype__001']) or 'rain' in str(row['days__preciptype__002'])) else 0, axis=1)
    df_clean['days_has_snow'] = df_clean.apply(lambda row: 1 if ('snow' in str(row['days__preciptype__001']) or 'snow' in str(row['days__preciptype__002'])) else 0, axis=1)

    # Drop original preciptype columns
    df_clean.drop(columns=['days__preciptype__001', 'days__preciptype__002'], inplace=True)

    # Reorder columns
    new_column_order = ['days__datetime', 'days_day_of_week', 'days__tempmax', 'days__tempmin', 'days__temp',
                        'days__feelslikemax', 'days__feelslikemin', 'days__feelslike',
                        'days_has_rain', 'days__precip', 'days__precipprob',
                        'days_has_snow', 'days__precipcover', 'days__snow', 'days__snowdepth',
                        'days__windspeed', 'days__cloudcover', 'days__visibility']
    df_clean = df_clean.reindex(columns=new_column_order)

    return df_clean
```
